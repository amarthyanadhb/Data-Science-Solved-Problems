1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.


```python
# Create a list of 5 integers
numbers = [10, 20, 30, 40, 50]
print("Original list:", numbers)

# 1. Append - add an element at the end
numbers.append(60)
print("After append(60):", numbers)

# 2. Extend - add multiple elements at the end
numbers.extend([70, 80])
print("After extend([70, 80]):", numbers)

# 3. Insert - add an element at a specific index
numbers.insert(2, 25)  # insert 25 at index 2
print("After insert(2, 25):", numbers)

# 4. Remove - remove first occurrence of a value
numbers.remove(40)
print("After remove(40):", numbers)

# 5. Pop - remove element at a specific index (default last)
popped = numbers.pop()  # removes last element
print("After pop():", numbers, "Popped element:", popped)

# 6. Index - find index of a value
index_30 = numbers.index(30)
print("Index of 30:", index_30)

# 7. Count - count occurrences of a value
count_20 = numbers.count(20)
print("Count of 20:", count_20)

# 8. Sort - sort the list in ascending order
numbers.sort()
print("After sort():", numbers)

# 9. Reverse - reverse the list
numbers.reverse()
print("After reverse():", numbers)

# 10. Clear - remove all elements
numbers.clear()
print("After clear():", numbers)
```

11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows



```python
import pandas as pd
import numpy as np

# 1. Create a small DataFrame manually with 10 rows
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah', 'Ian', 'Jack'],
    'Age': [25, 30, 35, 40, 28, 33, 29, 31, 27, 30],
    'City': ['Mumbai', 'Delhi', 'Kolkata', 'Mumbai', 'Delhi', 'Chennai', 'Kolkata', 'Mumbai', 'Delhi', 'Chennai'],
    'Salary': [50000, 60000, 70000, np.nan, 55000, 65000, 60000, np.nan, 58000, 62000]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)

# 2. Use .info() and .describe()
print("\nDataFrame info:")
print(df.info())

print("\nDataFrame description:")
print(df.describe())

# 3. Select 'Name' and 'City' columns
name_city = df[['Name', 'City']]
print("\nSelected columns (Name and City):\n", name_city)

# 4. Drop 'City' column
df_dropped = df.drop('City', axis=1)
print("\nDataFrame after dropping 'City' column:\n", df_dropped)

# 5. Fill missing values in 'Salary' column with mean
salary_mean = df['Salary'].mean()
df['Salary'].fillna(salary_mean, inplace=True)
print("\nDataFrame after filling missing Salary values:\n", df)

# 6. Remove any duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nDataFrame after removing duplicates:\n", df_no_duplicates)
```
